```python
# Reading the datasets
users = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
sessions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv' )
products = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv' )
transactions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
```

```python
# Converting the object datatype to datetime datatype - To acheive NaT result
users['Registered'] = pd.to_datetime(users['Registered'])
users['Cancelled'] = pd.to_datetime(users['Cancelled'])
users.head()
```

```python
# Converting the object datatype to datetime datatype - To acheive NaT result
sessions['SessionDate'] = pd.to_datetime(sessions['SessionDate'])
sessions.head()
```

```python
# importing the necessary packages and mapping the sqldf function as a global variable (pysqldf) 
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3 as db
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
# Performing a left join to keep all rows of transactions and matching rows from users
transactions.merge(users,on='UserID',how='left')
```

```python
# Using sub query and acheiving the result
# t.* alias will pick all rows from transactions and sub query will imply the condition. 
# 'Or' condition is used to consider the null values as well
pysqldf("select t.* from transactions t where t.UserID not in (select UserID from users) or t.UserID is null;")
```

```python
# Using the merge concept to acheive the result.
# Concept is like to full join and then a left join => concat and then drop the duplicates
trans_users = pd.concat([pd.merge(users,transactions,left_on='UserID',right_on='UserID',how='outer'),
           pd.merge(users,transactions,left_on='UserID',right_on='UserID',how='left')]).drop_duplicates(keep=False)
# Dropping the unwanted columns from the dataframe
trans_users.drop(['Gender','User','Registered','Cancelled'],axis=1,inplace=True)
# Obtaining the result in the desired format
trans_users[['TransactionID','TransactionDate','UserID','ProductID','Quantity']]
```

```python
# Performing a self join using the UserId column on the same transactions dataframe
transactions.merge(users,on='UserID',how='inner')
```

```python
# Performing a full outer join on transactions and users dataframes
transactions.merge(users,on='UserID',how='outer')
```

```python
# Join to get the sessions that occured on the same day using SQL query 
pysqldf("select u.*,s.SessionID,s.SessionDate from users u join sessions s on u.Registered = s.SessionDate;")
```

```python
# Join to get the sessions that occured on the same day using pandas merge function
result = pd.merge(users,sessions,left_on="Registered",right_on="SessionDate")
# Renaming the overlapping column
result.rename(columns={'UserID_x':'UserID'},inplace=True)
# Obtaining the output in the desired format
print(result[['UserID','User','Gender','Registered','Cancelled','SessionID','SessionDate']],"\n")
# To acheive the expected result given, we are using the loc method and isin method to filter the result
Exp_Res = result[['UserID','User','Gender','Registered','Cancelled','SessionID','SessionDate']]
Exp_Res.loc[Exp_Res['UserID'].isin(['2','4'])]
```

```python
# Dropping the NaN values from transaction
transactions_new = transactions.dropna()
# To get the possible pair of transactions, we do a full outer merge on the same transactins_new dataframe
pd.merge(transactions_new,transactions_new,left_on='UserID',right_on='UserID',how='outer')
```

```python
# First occuring transaction for each user using a left merge on users and transactions.
# To drop the dduplicates, we use a subset on 'User' and keep the first occurence
data = pd.merge(users,transactions,on='UserID',how='left').drop_duplicates(subset='User',keep='first')
data
```

```python
# Getting list of columns
my_columns = list(data.columns) 
my_columns
```

```python
# Dropping Nan values by defining a threshold value
list(data.dropna(thresh=int(data.shape[0] * .9), axis=1).columns)
```

```python
# Getting the list of columns in which null values are present
missing_info = list(data.columns[data.isnull().any()]) 
missing_info
```

```python
# looping through the columns in which null is present and using shape() method to count the missing data
print("Count of missing data\n")
for col in missing_info:
    num_missing = data[data[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
# Dividing the count by the total rows in the null value column = > we wil get the percentage of missing data 
print("Percentage of missing data\n")
for col in missing_info:
    percent_missing = data[data[col].isnull() == True].shape[0] / data.shape[0]
    print('percent missing for column {}: {} percent'.format( col, percent_missing))
```
